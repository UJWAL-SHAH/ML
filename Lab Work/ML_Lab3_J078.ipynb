PROCEDURE:

1) We divide the countries into different continents. We find the Mean of "Total" immigrated population of the "Individual Continents". We also find Standard Deivation of Immigrated population of "Individual" Continents.

2) Now we  find Mean Immigration of each countries.

3) Now we do study for each continent as countries in their respective continents show similiar trends easily.

4) Now we study that countries in the continent whose mean is between (Mean of Asia  ± Standard deviation of Asia).

5) So now the respective countries graph will be plotted in time series so as to look into similarity between trends & the outliers will automatically be removed.

6)This process will be repeated for all other continents also
